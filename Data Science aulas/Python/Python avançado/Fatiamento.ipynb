# Exemplo 1: Ler a nota turma: 1° Bimestre.


- Estrutura de Dados: Numeric Python (numpy)

EXEMPLO: Vetor Nota do 1° Bimestre.
- Ler notas [0, 10] dos 50 Alunos: Valor Final
  - Preencher os Dados de um vetor:
  - 1: Ler do usuário.
  - 2: Leitura de Arquivos: csv, json, dat, html, ...
  - 3: Simulação de dados: random

```
import numpy
indice = 0
TAMANHO = 3  # Tamanho máximo: 50 Alunos
bimestre1 = numpy.zeros(TAMANHO)
while (indice < TAMANHO):
  print(f'Nota [0, 10] do aluno {indice + 1}: ')
  bimestre1[indice] = float(input())
  if (bimestre1[indice] < 0 or bimestre1[indice] > 10):
    print('Erro na entrada dos dados. Escolha novamente.')
  else:
    indice += 1
```

# Medidas de dispersão:

```
print('Medidas de dispersão:')
print(f'Mínimo: {dataset.min()} Gramas.')
print(f'Máximo: {dataset.max()} Gramas.')
print(f'Amplitude: MAX - MIN: {dataset.max() - dataset.min()} Gramas.')
print(f'Variância: {dataset.std() ** 2: .2f} Gramas.')
print(f'Desvio Padrão amostral: {dataset.std(): .2f} Gramas.')
print(f'Coeficiente de Variação (CV): {dataset.std() / dataset.mean() * 100: .2f} %.')
```

# Fatiamento: slicing

```
import numpy
media = 5.0
desvio = 1.8
alunos = 50
bimestre1 = numpy.random.normal(media, desvio, alunos)
bimestre1
# Sem Formatação 
```

```
# com formatação:
bimestre1.sort()
print('Notas do Bimestre 1:')
for index in range (alunos):
  print(f'Nota do Aluno {index + 1}: {bimestre1[index]: .1f}')
```

# Fatiamento: Ordem Crescente

-  Dados um dataset temos:
``subdataset = dataset[INICIO:FIM:ITERAÇÃO]``
- INICIO   = 0 (Default)
- FIM.     = LEN (Tamanho)   (Default: não incluso)
- ITERAÇÃO = 1 (Default)

```
len(bimestre1[0:50:1])
```

# Fatiamento: Ordem Crescente: Alterar Final

```
subdataset = bimestre1[:alunos//4]
len(subdataset)
```

# Fatiamento: Ordem Crescente: Alterar Inicio

```
subdataset = bimestre1[3 * alunos // 4:]
len(subdataset)
```

# Fatiamento: Ordem Crescente: Alterar Inicio e Fim

```
quartil2 = bimestre1[alunos // 4: 3 * alunos // 4 ]
print('Notas do Bimestre 1: 2 Quartil')
for index in range (alunos // 4, 3 * alunos // 4):
  print(f'Nota do Aluno {index + 1}: {bimestre1[index]: .1f}')
```

# Fatiamento: Ordem Decrescente: Alterar Inicio e Fim

- Dados um dataset temos (Ordem Decrescente):
 
    ``subdataset = dataset[INICIO:FIM:ITERAÇÃO]``

- INICIO   = -1
- FIM.     = -1 * len (dataset) - 1 (fim: não incluso)
- ITERAÇÃO = -1 

```
# com formatação:
print('Notas do Bimestre 1:')
for index in range (alunos):
  print(f'Nota do Aluno {index + 1}: {bimestre1[index]: .1f}')
# sem formatação: Sem Modificação do dataset.
bimestre1[-1: -1 * len(bimestre1) - 1:-1]
```
