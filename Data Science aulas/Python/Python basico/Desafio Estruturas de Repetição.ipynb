## Fazer dois (2) exerc√≠cios de Repeti√ß√£o da lista usando o comando **FOR + IN RANGE** (a sua escolha);

1) Escrever um algoritmo para exibir os m√∫ltiplos de 3 compreendidos no intervalo: [3 100].

```python
print("Os multiplos de 3 sao : ")
for m in range (3,100):
  if m % 3 == 0:
    print(f'numero: {m}')
```

2) Escrever um algoritmo para exibir os m√∫ltiplos de 11, a soma e a m√©dia dos m√∫ltiplos de 11, em ordem
decrescente (inversa), compreendidos entre o intervalo: [200 100].


```python
s = 0
for c in range(200,100,-1): 
  if c % 11 == 0:
    s+=c
    print(f"{c}")
print(f"soma {s}")
print(f"media {s/2}")
```

5) Escreva um algoritmo que leia de 10.000 habitantes de uma pequena cidade se est√° empregado ou n√£o e
exiba em porcentagem a quantidade de empregados e desempregados desta pequena cidade.

```python
hab = 0
e = 0
d = 0
while hab < 10000:
  print('vc esta empregado?')
  print('[1] para Sim')
  print('[2] para N√£o')
  print('[0] para acabar pesquisa')
  opcao = int(input(f'resposta '))
  if hab == 10000:
    print('obrigado por responder o questionario')
    break
  elif opcao == 0:
    print('obrigado por responder o questionario')
    break
  elif opcao == 1:
    e += 1
  else:
    d += 1
  empregados = e / (e + d) * 100
  desempregados = d / (e + d) * 100
  print(f'empregados: {empregados}%.')  
  print(f'desempregados: {desempregados}%.')
  hab += 1
```

6) Escreva um algoritmo que leia o sal√°rio em reais (R$) de 1000 clientes de um shopping e exiba na tela,
em porcentagem, a divis√£o dos clientes por tipo: A, B ou C, conforme a seguir:

Declarar o Sal√°rio M√≠nimo (SM: Pesquisar no Google)

*   MENOR que o Segredo da porta.

```python
import numpy as np
```

```python
# ESQUELETO DO PROGRAMA:
while True:
  print("MENU: ")
  print("Escolha: [1] para jogar ou [0] para sair: ")
  resposta = int(input())
  if (resposta == 0):
    print("Fim do Jogo.")
    break
  elif (resposta != 1):
    print("Erro. Escolha de novo")
  else:
    segredo = np.random.randint(1, 101, 1)
    print(f"Segredo: {segredo[0]}")
    
    # FAZER O JOGO AQUI!  # No m√°ximo 7 chances antes da morte!
  tentativas = 0 
  while True:    
    respostag = int(input('adivinhe um numero que esteja entre 0 e 100 ? '))
    tentativas+=1        
    while respostag != segredo[0] or respostag == segredo[0]:
      if tentativas < 8 and respostag < segredo[0]:
        print(f'tente novamente o numero que procura e MAIOR que esse NUMERO {respostag} vc tem {tentativas} ')
        break
      elif tentativas < 8 and respostag > segredo[0]:
        print(f'tente novamente o numero que procura e MENOR que esse NUMERO {respostag} vc tem {tentativas}')
        break
      elif respostag == segredo[0]:
        print(f'voce acertou o numero e {segredo[0]}')
        break
      else:
        print(f'voce morreu')
        break
```
