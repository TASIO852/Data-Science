# EXEMPLO 1: Distância entre os pontos cartesianos: $\ P$ e $\ Q$.

Distancia entre $ \ P(Px, Py)$ e $ \ Q(Qx, Qy)$ = $ \sqrt{(P_x - Q_x)^2 + (P_y - Q_y)^2}$

```python
def distanciaEntrePontos(Px=0, Py=0, Qx=0, Qy=0):
  """
  AJUDA DA FUNÇÃO: Cálculo entre os pontos cartesianos:
  P(Px, Py) e Q(Qx, Qy).
  Por default: P(0, 0) e Q(0, 0).
  """
  distancia = ((Px - Qx) ** 2 + (Py - Qy) ** 2) ** 0.5
  return distancia
```

```python
# Chamada 1: Usando todos os parâmetros da função:
distancia = distanciaEntrePontos(3, 0, 0, 4)
print(f'Distância entre os pontos P({3: .1f}, {0: .1f}) e Q({0: .1f}, {4: .1f}) = {distancia: .1f}')
```

```python
# Chamada 2: Omitindo um (1) parâmetro: 4° Parâmetro.
distancia = distanciaEntrePontos(0, 3, 4)
print(f'Distância entre os pontos P({0: .1f}, {3: .1f}) e Q({4: .1f}, {0: .1f}) = {distancia: .1f}')
```

```python
# Chamada 3:  Omitindo três (3) parâmetros: 2°, 3° e 4° Parâmetros.
distancia = distanciaEntrePontos(-5)
print(f'Distância entre os pontos P({-5: .1f}, {0: .1f}) e Q({0: .1f}, {0: .1f}) = {distancia: .1f}')
```

```python
# Chamada 4:  Omitindo todos os parâmetros.
distancia = distanciaEntrePontos()
print(f'Distância entre os pontos P({0: .1f}, {0: .1f}) e Q({0: .1f}, {0: .1f}) = {distancia: .1f}')
```

```python
# Chamada 4:  Omitindo 4º parâmetros.
distancia = distanciaEntrePontos(-1, -3, 4)
print(f'Distância entre os pontos P({-1: .1f}, {-3: .1f}) e Q({4: .1f}, {0: .1f}) = {distancia: .1f}')
print(f'Distância entre os pontos P({-1: .1f}, {-3: .1f}) e Q({4: .1f}, {0: .1f}) = {distanciaEntrePontos(-1, -3, 4): .1f}')
```

# EXEMPLO 2: 

Fatorial:

Fatorial(n):

n = 0 ou 1     Fat(0 ou 1) = 1

n >= 2         Fat(n >= 2) = 1 * 2 * 3 * ...* n

```python
def Fatorial(N):
  """
  AJUDA:
  Fatorial:
  Fatorial(n):
  n = 0 ou 1 Fat(0 ou 1) = 1
  n >= 2 Fat(n >= 2) = 1 * 2 * 3 * ...* n
  """
  if N == 0 or N == 1:
    return 1
  fat = 1
  for contador in range (1, N + 1):
    fat = fat * contador
  return fat
```

# EXEMPLO 3: Função para estabelecer os índices de Saúde.

# CÉLULA 1: Declaração da Função: Índices de Saúde.

```python
def indices_de_saude(massa, altura, medidaCintura, medidaQuadril):
  """
  AJUDA DA FUNÇÃO (DOCSTRING):
  Parâmetros de entrada:
  _massa: em quilogramas;
  _altura: em quilogramas;
  _medidaCintura: em centímetros;
  _medidaQuadril: em centímetros;

  Índices de Saúde implementa como objetivo determinar os parâmetros de 
  referência de saúde com os seguintes indicadores, a saber:
  -> RCQ: Relação Cintura-Quadril = _medidaCintura / _medidaQuadril * 100 (%)
  -> IMC: Índice de Massa Corpórea = Massa / (altura ** 2)
  -> Classificação por IMC, sendo:
      <> "MAGREZA": IMC < 18.5
      <> "PESO NORMAL": IMC < 25
      <> "SOBREPESO": IMC < 30
      <> "OBESIDADE GRAU I": IMC < 35
      <> "OBESIDADE GRAU II": IMC < 40
      <> "OBESIDADE GRAU III": IMC >= 40
  """
  imc = massa / (altura ** 2)
  rcq = medidaCintura / medidaQuadril * 100
  return imc, rcq
```

# EXEMPLO 4: 

Função Número Perfeito:



image.png

```python
def perfectNumber(number):
  """
  Em Matemática, um número perfeito é um número natural 
  para o qual a soma de todos os seus divisores naturais próprios (excluindo ele mesmo) 
  é igual ao próprio número.[1]Por exemplo, o número 28 é , pois: 
  28 = 1 + 2 + 4 + 7 + 14
  Todo número perfeito é um número triangular, bem como um número hexagonal.
  """
  sum = 1
  for counter in range(2, (number // 2) + 1):
    if (number % counter == 0):
      sum = sum + counter
  if sum == number:
    return True  # True: Sim, é perfeito
  else:
    return False # False: Não, é perfeito
```

```python
# REUSABILIDADE:
print('Todos os Números Perfeitos entre [1, 100000]: ')
for i in range(1, 1000):
  if perfectNumber(i):
    print(f'Numero Perfeito: {i}')
```

# EXEMPLO 5: 

Criar a função: Equação do 2º Grau:

image.png

```python
def equacao2(A, B, C):    # Protótipo: se usar: _ (underline) Variavel Local
  #Fazer a Lógica da Função Aqui:
  delta = B ** 2 - 4 * A * C
  if delta < 0:
    resposta = False      # NÃO TEM RAÍZES NO REAIS
    raizDelta = False
    X1 = False
    X2 = False
  else:
    resposta = True       # TEM RAÍZES NO REAIS
    raizDelta = delta ** 0.5
    X1 = (- B + raizDelta) / (2 * A) 
    X2 = (- B - raizDelta) / (2 * A) 
  return delta, raizDelta, X1, X2, resposta
```

```python
"""
CÉLULA DO PROGRAMA PRINCIPAL
Usar em 100 Equações diferentes, lidos os coeficiente: A, B e C
do usuárioio.
"""
contador = 0                    # VALOR INICIAL
while contador < 100:           # VALOR FINAL
  print(f'Equação {contador + 1}: ')
  A = float(input('Coeficiente A: '))
  if (A == 0):
    print('Erro no coeficiente A. Escolha de novo.')
  else:
    B = float(input('Coeficiente B: '))
    C = float(input('Coeficiente C: '))
    delta, raizDelta, X1, X2, resposta = equacao2(A, B, C)
    print(f'Delta: {delta: .2f}')
    if resposta == False:
      print('NÃO TEM RAÍZES NO REAIS')
    else:
      print(f'Raiz de Delta: {raizDelta: .2f}')
      print(f'X1: {X1: .2f}')
      print(f'X2: {X2: .2f}')
      contador += 1             # ITERAÇÃO
```

# 2. Crie uma função e seu protótipo que receba como argumento o preço de uma mercadoria e seu reajuste em porcentagem, retorne o preço reajustado e o reajuste calculado em reais. Após isso, use a função no programa principal e exiba o preço reajustado e o reajuste de 100 mercadorias informadas pelo usuário.**

```python
def mercadoria(preco,reajuste):
  if preco >= 100:
    reajuste_porcentagem = reajuste / 100
    preco_reajustado = (preco * reajuste_porcentagem) + preco
    reajusteReais = preco_reajustado - preco
    return reajusteReais,preco_reajustado
  elif preco <= 100:
    reajuste_porcentagem = reajuste / 100
    preco_reajustado = preco - (preco * reajuste_porcentagem)
    reajusteReais = preco_reajustado - preco
    return reajusteReais,preco_reajustado
  else:
    return False
```

Hora de entrada (HH:MM): Exemplo: 13:37 -
Hora de saída (HH:MM): Exemplo: 18:41 = (HH:MM): Exemplo: 04:64 
→ 5 Horas de permanência no estacionamento.

Preço: R$ 7.00 por hora.


```python
def parquimetro(HE,ME,HS,MS,preco):
  permH = (HS-HE)*60
  permM = MS-ME 
  permanencia = (permH + permM) / 60
  preco = permanencia * 7 
  return permanencia,preco
```

Na célula do programa principal, use a função várias vezes, criando um Menu para o usuário.


```python
def triangulo(a,b,c):
  if (a + b < c) or (a + c < b) or (b + c < a):
    return True # nao forma triangulo
  else:
    return False # forma triangulo
```

```python
# DIZER o tipo do triangulo ja sabendo se pode ou nao formar um
while True:
  a = int(input('coloque as medidas do triangulo: '))
  b = int(input('coloque as medidas do triangulo: '))
  c = int(input('coloque as medidas do triangulo: '))
  if triangulo(a,b,c) == True:
    print('Essas medidas nao forman um triangulo')
  elif a == b == c:
    print('Equilatero')
  elif a==b or a==c or b==c:
    print('Isósceles')
  else:
    print('Escaleno')
  break
```
