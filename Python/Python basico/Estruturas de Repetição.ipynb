## Definição de **Estruturas de Repetição de Dados**: 

$\longmapsto$ São usadas justamente para repetir um **BLOCO DE INSTRUÇÕES** (conjunto de linhas de códigos) sucessivas vezes

até atingir uma determinada **CONDIÇÃO** **(Lógica)** de parada.

$\longmapsto$ **BLOCO DE INSTRUÇÕES** (conjunto de linhas de códigos): Na linguagem de programação Python, o bloco é determinado por **espaços em branco razão de múltiplos de 4: 4, 8, 12, 16, 20, ....;** para cada novo bloco de controle.

```
while True:
  print('MENU DE OPÇÕES PARA O USUÁRIO:')
  print('OPÇÃO 1: Calcular a Equação do 2° Grau.')
  print('OPÇÃO 0: Sair do Programa.')
  opcao = int(input('Qual é a sua Opção: '))
  if opcao == 0:
    print('FIM DO PROGRAMA! TCHAU!!!')
    break
  elif opcao != 1:
    print('Opção incorreta do Menu. Escolha de novo, por favor!!!')
  else:
    A = float(input('Coeficiente A: '))
    B = float(input('Coeficiente B: '))
    C = float(input('Coeficiente C: '))
    if A == 0 or (B == 0 and C == 0):
      print('Não é uma Equação do 2° Grau.')
    else:
      delta = B ** 2 - 4 * A * C
      print(f'Valor de Delta: {delta: .2f}')
      if delta < 0:
        print('Não tem Raízes no conjunto dos números reais.')
      elif delta == 0:
        X1 = -B / (2 * A)
        print(f'Raiz X1 = X2 = {x1: .2f}')
      else:
        raizDelta = delta ** 0.5
        print(f'Valor da Raíz de Delta: {raizDelta: .2f}')
        X1 = (-B + raizDelta) / (2 * A)
        X2 = (-B - raizDelta) / (2 * A)
        print(f'Raiz X1 = {X1: .2f}')
        print(f'Raiz X2 = {X2: .2f}')
```

## **Estrutura de Repetição com teste condicional**: Comando WHILE.

### $\longmapsto$ **COMANDO WHILE**:

while (condição):
  
$\quad$ // AÇÃO 1

$\quad$  // AÇÃO 2

$\quad$  // AÇÃO 3

 $\quad$ // AÇÃO ...

```
"""
PADRÃO:
-> Parâmetros da Repetição:
-> Valor Inicial
-> Valor Final
-> Iteração
contador = Valor Inicial
while contador <= Valor Final:
  # Ação 1
  # Ação 2
  # Ação 3
  ...
  # Iteração:
  contador = contador + passo
"""
print('PADRÃO DO WHILE PARA CONTAGEM: PROGRESSIVA')
contador = 1
while contador <= 10:
  print(f'Contagem progressiva: {contador}')
  contador = contador + 1 # ITERAÇÃO: Passo = 1
print('\nPADRÃO DO WHILE PARA CONTAGEM: REGRESSIVA')
contador = 10
while contador >= 1:
  print(f'Contagem Regressiva: {contador}')
  contador = contador - 1 # Passo = -1
```

```
"""
EXEMPLO 1:
Ler a nota[0, 10] de 50 alunos (VALOR FINAL) de uma turma e exibir sua a média geral.
Média = Soma as Notas de todos / Alunos
"""
alunos = 0
soma   = 0
while alunos < 5:
  nota = float(input(f'Nota [0, 10] do aluno {alunos + 1}: '))
  # TRATAMENTO DE ERRO.
  if nota < 0 or nota > 10:
    print('Erro na digitação da nota. Escolha novamente.')
  else:
    soma = soma + nota
    alunos = alunos + 1
media = soma / alunos
print(f'Média da Turma: {media: .1f}')
```

```
"""
EXEMPLO 2:
Ler a opinião de 1000 alunos universitários (VALOR FINAL) sobre a seguinte questão:
Você acha importante Saber Programar? 1: SIM ou 0: NÃO

Exiba o resultado da pesquisa em porcentagem(%)

-> Como calcular a Porcentagem?
% Parte = Parte / Total * 100
% sim = sim / (sim + nao) * 100
% nao = nao / (sim + nao) * 100
"""
alunos = 0                 # VALOR INICIAL
sim = 0
nao = 0
while alunos < 5:          # VALOR FINAL
  print('MENU:')
  print('Você acha importante Saber Programar? 1: SIM ou 0: NÃO')
  resposta = int(input())
  if resposta != 1 and resposta != 0:
    print('Resposta incorreta. Escolha novamente.')
  else:
    if resposta == 0: # Resposta Não
      nao = nao + 1
    else:
      sim = sim + 1
    print(f'Porcentagem de Não: {nao / (nao + sim) * 100: .1f}%')
    print(f'Porcentagem de Sim: {sim / (nao + sim) * 100: .1f}%')
    #alunos = alunos + 1
    alunos += 1               # ITERAÇÃO
```

```
"""
EXEMPLO 3:
Ler o voto de 100 moradores para síndico do prédio.
Os cadidatos são:

CRIAR UM MENU DE OPÇÕES:

1: Fulano de Tal
2: Beltrano de Tal
3: Cicrano de Tal

Exiba o resultado da votação em porcentagem(%)
fulano   = ?
beltrano = ?
cicrano  = ?
""" 
moradores = 0
fulano    = 0
beltrano  = 0
cicrano   = 0
while moradores < 100:
  print('MENU DE OPÇÕES - Escolha: ')
  print('1: Fulano de Tal')
  print('2: Beltrano de Tal')
  print('3: Cicrano de Tal')
  voto = int(input(f'Voto do Morador {moradores + 1}: '))
  if voto <= 0 or voto >= 4:
    print('Erro no voto. Escolha de novo!')
  else:
    # moradores = moradores + 1
    moradores += 1
    if voto == 1:
      fulano += 1
    elif voto == 2:
      beltrano += 1
    else:
      cicrano = cicrano + 1
    print('Resultados parciais da votação: ')
    print(f'Votos de Fulano. : {fulano / moradores * 100: .1f} %.')
    print(f'Votos de Beltrano: {beltrano / moradores * 100: .1f} %.')
    print(f'Votos de Cicrano : {cicrano / moradores * 100: .1f} %.')
```

## **Estrutura de Repetição com interrupção (comando break)**: Comando WHILE + BREAK.

### $\longmapsto$ **COMANDO WHILE(True) + BREAK**:

É mais adequado para problemas onde não é informado o VALOR FINAL.

while (True):
  
$\quad$ // AÇÃO 1

$\quad$  // AÇÃO 2

$\quad$  // AÇÃO 3

 $\quad$ // AÇÃO ...

$\quad$if (valor == PARAR):

$\quad$$\quad$break

```
"""
EXEMPLO 1:
Ler a opinião de VÁRIOS alunos universitários sobre a seguinte questão:
Você gosta de programação de computadores? 1: SIM ou 2: NÃO
Exiba o resultado da pesquisa em porcentagem(%), 
parando a pesquisa quando o usuário digitar VALOR = ZERO. 
(CONDIÇÃO DE PARAR O PROGRAMA)
"""
alunos = 0
sim = 0
nao = 0
print('Você gosta de programação de computadores?')
while True:
  print('MENU DE OPÇÕES: ')
  print('1: para Sim')
  print('2: para Não')
  print('zero: para sair do programa')
  opcao = int(input(f'Opção do Aluno {alunos + 1}: '))
  if opcao == 0:
    print('Fim do Programa')
    break
  elif opcao != 1 and opcao != 2:
    print('Erro na opção. Escolha novamente!!!')
  else:
    if opcao == 1:
      sim += 1
    else:
      nao += 1
    print('Resultado parcial da Pesquisa: ')
    ps = sim / (sim + nao) * 100
    print(f'Porcentagem de Sim: {ps: .1f}%.')
    pn = nao / (sim + nao) * 100
    print(f'Porcentagem de Não: {pn: .1f}%.')
    alunos += 1     # Iteração
```

```
"""
EXEMPLO 2: WHILE + BREAK

Faça um algoritmo que leia VÁRIOS números reais (um por um) e 
exiba na tela a média dos positivos e negativos escolhidos do usuário.

PARE (break) o programa quando o usuário digitar VALOR = ZERO. 
(CONDIÇÃO DE PARAR O PROGRAMA)

OBJETIVO:
Média dos Positivos = Soma dos Positivos / Quantidade de Nº Positivos lidos do usuário
mp = sp / qp
Média dos Negativos = Soma dos Negativos / Quantidade de Nº Negativos lidos do usuário 
mn = sn / qn
"""
sp = 0
qp = 0
sn = 0
qn = 0
while True:
  print('MENU:')
  print('Escolha um número real qualquer ou ')
  print('Escolha ZERO para sair.')
  numero = float(input('Entre com sua Opção: '))
  if numero == 0:
    print('Fim do programa.')
    break
  elif numero > 0: 
    sp = sp + numero
    qp = qp + 1
  else:
    sn += numero
    qn += 1
if qp != 0:
  print(f'Média dos positivos: {sp / qp: .1f}')
else:
  print('Sem valores positivos lidos do usuário.')
if qn == 0:
  print('Sem valores negativos lidos do usuário.')
else:
  print(f'Média dos Negativos: {sn / qn: .1f}')
```
