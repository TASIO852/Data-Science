```
A = float(input('Coeficiente A: '))
B = float(input('Coeficiente B: '))
C = float(input('Coeficiente C: '))
if A == 0 or (B == 0 and C == 0):
  print('Não é uma Equação do 2° Grau.')
else:
  delta = B ** 2 - 4 * A * C
  print(f'Valor de Delta: {delta: .2f}')
  if delta < 0:
    print('Não tem Raízes no conjunto dos números reais.')
  elif delta == 0:
    X1 = -B / (2 * A)
    print(f'Raiz X1 = X2 = {x1: .2f}')
  else:
    raizDelta = delta ** 0.5
    print(f'Valor da Raíz de Delta: {raizDelta: .2f}')
    X1 = (-B + raizDelta) / (2 * A)
    X2 = (-B - raizDelta) / (2 * A)
    print(f'Raiz X1 = {X1: .2f}')
    print(f'Raiz X2 = {X2: .2f}')
```

Escrever um algoritmo em Python que leia a comprimento de uma (1) barra em centímetro e exiba na tela sua dimensão em:

*   pés

## **Exemplo 1**: Seleção Simples (IF)

Escrever um algoritmo em Python que leia o Preço (R$) e a quantidade (unidades) de uma (1) Mercadoria e 

exiba o total a pagar. Entranto, se o total a pagar ultrapassar R$ 1000.00 dê 3% desconto ao cliente.

```
"""
SELEÇÃO SIMPLES: IF (SE)

if (CONDIÇÃO LÓGICA):
  // BLOCO DE CONTROLE:
  // AÇÃO 1
  // AÇÃO 2
  // AÇÃO 3
  // AÇÃO ...
"""
# ALGORITMO = ENTRADA + PROCESSAMENTO + SAÍDA
# ENTRADA
preco = float(input('Entre com o Preço: R$ '))
quantidade = int(input('Entre com a quantidade (em Unidades): '))
# PROCESSAMENTO: Cálculos Aritméticos + Lógicos
total = preco * quantidade
if total > 1000:
  total = total -0.03 * total
  # total = 0.97 * total
# SAÍDA (Objetivo)
print(f'Total a pagar: R$ {total: .2f}')
```

## **Exemplo 2**: Seleção Composta (IF - ELSE)

Escrever um algoritmo que leia um(1) número: POSITIVO e exiba na tela:

$\longmapsto$ O quadrado do número, se o número for PAR E NÃO MÚLTIPLO DE 5 ou

$\longmapsto$ A raiz quadrada do número, caso contrário.

```
"""
SELEÇÃO SIMPLES: IF - ELSE

if (CONDIÇÃO LÓGICA):
  // BLOCO DE CONTROLE: IF
  // AÇÃO 1
  // AÇÃO 2
  // AÇÃO 3
  // AÇÃO ...
else:
  // BLOCO DE CONTROLE: ELSE
  // AÇÃO 1
  // AÇÃO 2
  // AÇÃO 3
  // AÇÃO ...
"""
numero = int(input('Entre com um valor INTEIRO e POSITIVO: '))
if numero > 0:
  # SE Numero for PAR e Não multiplo de 5
  if numero % 2 == 0 and numero % 5 != 0:
    print(f'O quadrado do Numero ({numero})= {numero ** 2}')
  else:
    print(f'A raiz do Numero ({numero})= {numero ** 0.5: .2f}')
else:
  print('ERRO: O número deve ser POSITIVO!')
```

## **Exemplo 4**: Seleção Encadeada (IF - ELIF - ELSE)

```
"""
PADRÃO IF- ELIF - ELSE
if Condição 1:
  // Bloco if
elif Condição 2:
  // Bloco elif
else: 
  // Bloco else
 ENUNCIADO DO EXEMPLO: Notas da UVV
 Escrever um algoritmo que leia a nota do 1º e 2º Bimetre de um (1) aluno e 
 exibir a sua média semestral com sua classificação (Status), a saber:
-> Nota Semestral: [0, 3[ => Aluno Status: Reprovação.
-> Nota Semestral Final: [3, 7[ => Aluno Status: Prova Final.
    -> Ler a nota da Prova Final e exibir o Status:
       -> [0, 5[: Status: Reprovação.
       -> [5, 10]: Status: Aprovação.
-> Nota Semestral Final: [7, 10] => Aluno Status: Aprovação.
"""
```

```
bimestre1 = float(input('Qual foi sua note do 1° bimestre ? '))
bimestre2 = float(input('Qual foi sua note do 2° bimestre ? '))

media = (bimestre1 + bimestre2) / 2

if media <= 3.0:
  print(f'Voce estar reprovado ')
else:
  if media > 3.0 and media < 7:
    print(f'Voce precisa fazer prova final porque sua media foi de {media}')


  if media > 3.0 and media < 7:
    provaf = float(input('Quanto voce tirou na prova final ? '))
    if provaf <= 5.0:
      print(f'Nao foi dessa vez amigao talvez ano que vem')
    elif provaf > 5.0: 
      print(f'Parabens voce passou {media}  ')
    else: 
      print(f'Parabens voce passou direto sua media foi de {media}')

#codigo do alesssandro !!!

# TRATAMENTO DE EXCEÇÃO:
try:
  B1 = float(input('Entre com a nota [0, 10] 1º bimestre: '))
  B2 = float(input('Entre com a nota [0, 10] 2º bimestre: '))
  # TRATAMENTO DE ERRO:
  if ((B1 < 0 or B1 > 10) or (B2 < 0 or B2 > 10)):
    print('Erro na digitação.')
  else:
    MS = (B1 + B2) / 2
    print(f'Média semestral: {MS: .1f}')
    if MS >= 7.0:
      print('STATUS: APROVAÇÃO')
    elif MS >= 3.0:
      print('STATUS: PROVA FINAL')
      PF = float(input('Entre com a nota [0, 10] da Prova Final: '))
      # TRATAMENTO DE ERRO:
      if (PF < 0 or PF > 10):
        print('Erro na digitação.')
      else:
        MS = (MS + PF) / 2
        print(f'Média semestral: {MS: .1f}')
        if MS >= 5:
          print('STATUS: APROVAÇÃO')
        else:
          print('STATUS: REPROVAÇÃO')
    else:
      print('STATUS: REPROVAÇÃO')
except Exception as erro:
  print(f'Erro: {erro}')
```
