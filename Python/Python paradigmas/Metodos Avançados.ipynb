# Ler a nota turma: 1° Bimestre.

```
ALUNOS = 5
bimestre1 = np.zeros(ALUNOS)
bimestre1
```

```
"""
EXEMPLO: Vetor Nota do 1° Bimestre.
-> Ler notas [0, 10] dos 50 Alunos: Valor Final
-> Objetivo: Exibir a média da turma 1º Bimestre:
"""
indice = 0
while (indice < ALUNOS):
  # TRATAMENTO DE EXCEÇÃO
  try:
    print(f'Nota [0, 10] do Aluno {indice + 1}: ')
    bimestre1[indice] = float(input())
    # TRATAMENTO DE ERRO:
    if (bimestre1[indice] < 0 or bimestre1[indice] > 10):
      print('Erro na digitação da nota. Escolha de novo.')
    else:
      #indice = indice + 1
      indice += 1 
  except Exception as erro:
    print(f'Erro: {erro}')
```

# Ler e exibir uma Lista: Notas do 1° Bimestre.

```
"""
EXEMPLO: Vetor Nota do 1° Bimestre.
-> Ler notas [0, 10] dos 50 Alunos: Valor Final
"""
import numpy                  # Biblioteca Numeric Python: numpy
bimestre1 = numpy.zeros(5)
indice = 0
while indice < 5:
  print(f'Nota [0, 10] - Aluno {indice + 1}: ')
  bimestre1[indice] = float(input())
  # TRATAMENTO DE ERRO: NOTA ] -inf, 0[ ou ]10, inf[
  if bimestre1[indice] < 0 or bimestre1[indice] > 10:
    print('Erro na digitação da nota. Escolha novamente o valor.')
  else:
    indice = indice + 1
print('Lista de Notas do 1° Bimestre: ')
for indice in range(len(bimestre1)):      # Comando len (lenght): tamanho
  print(f'Aluno {indice + 1} = {bimestre1[indice]}')
print(f'Lista de Notas do 1° Bimestre >= Media da Turma: {numpy.mean(bimestre1):.1f}')
for indice in range(len(bimestre1)):
  if (bimestre1[indice] > bimestre1.mean()):
    print(f'Aluno {indice + 1} = {bimestre1[indice]}')
```

# Metodos / Análise Estatística dos dados

```
"""
EXEMPLO: Vetor Nota do 1° Bimestre.
-> Ler notas [0, 10] dos 50 Alunos: Valor Final
"""
import numpy                  # Biblioteca Numeric Python: numpy
bimestre1 = numpy.zeros(50)
indice = 0
while indice < len(bimestre1):
  print(f'Nota [0, 10] - Aluno {indice + 1}: ')
  bimestre1[indice] = float(input())
  # TRATAMENTO DE ERRO: NOTA ] -inf, 0[ ou ]10, inf[
  if bimestre1[indice] < 0 or bimestre1[indice] > 10:
    print('Erro na digitação da nota. Escolha novamente o valor.')
  else:
    indice = indice + 1

# Dados Estatísticos da Turma
print(f'Média da Turma: {numpy.mean(bimestre1): .1f}')
print(f'Média da Turma: {bimestre1.mean() : .1f}')
print(f'Menor nota da Turma: {bimestre1.min() : .1f}')
print(f'Menor nota da Turma: {numpy.min(bimestre1) : .1f}')
print(f'Maior nota da Turma: {bimestre1.max() : .1f}')
print(f'Soma das notas da Turma: {bimestre1.sum() : .1f}')
print(f'Desvio Padrão nota da Turma: {bimestre1.std() : .1f}')

# Ordem
bimestre1Ordem = numpy.sort(bimestre1)
print('Lista de Notas do 1° Bimestre (Ordem Crescente): ')
for indice in range(len(bimestre1Ordem)):     # Método len = TAMANHO.
  print(f'Aluno {indice + 1} = {bimestre1Ordem[indice]}')
```

```
for indice, preco in enumerate(carrinho_de_compra):
  carrinho_de_compra[indice] = preco * 0.97
```

# INDEXACÃO BOOLEANA

Consiste em aplicar diretamente ao dataset (ou subdataset) operações: 

*   Relacionais: >, >=, <, <=, == e !=  e/ ou

*   Lógicos: Conjunção(and : &), Disjunção (or : |) ou Negação (not: ~)

como resultado, temos um vetor (ndarray) de valores booleanos (False = 0 ou True = 1)

```
"""
Exemplo 1:
Dado um conjunto de 100 senhas de 4 dígitos, exiba a quantidade de senhas 
que possuem como criptografia as características de serem pares e múltiplos 13.
"""
start = 1000
stop  = 9999
size = 100
senhas = np.random.randint(start, stop + 1, size)
```

```
senhas.sort()
senhas
```

Exiba a quantidade de senhas que possuem como criptografia as características de serem pares e múltiplos 13.

Como? Por Indexação Booleana

```
"""
Exiba a quantidade de senhas que possuem 
como criptografia as características de serem pares e múltiplos 13.
Como? Por Indexação Booleana
"""
                  #       PAR       and   MÚLTIPLAS DE 13   
index_bool_array = (senhas % 2 == 0) & (senhas % 13 == 0)
index_bool_array
```

```
print(f'Quantidade de senhas que possuem como criptografia: {sum(index_bool_array)}')
```

```
# Usando o vetor: indexbool
senhas_criptografadas = senhas[index_bool_array]
print('Lista de senhas sugeridas: ')
for index, value in enumerate(senhas_criptografadas):
  print(f'Sugestão de senha {index + 1}: {value}')
```

```
# Usando o vetor: indexbool
senhas_nao_criptografadas = senhas[~index_bool_array]
print('Lista de senhas não criptografadas: ')
for index, value in enumerate(senhas_nao_criptografadas):
  print(f'Sugestão de senha {index + 1}: {value}')
```
