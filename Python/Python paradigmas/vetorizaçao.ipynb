```python
"""
-> TABELA (Matriz): Modelagem do problema
-> LINHAS: Eventos (pessoas, mercadorias, ...)
-> COLUNAS: Variáveis
-> TABELA = LINHAS x COLUNAS = 100 x 4 = 400 Dados
-> COLUNA 0: MATRICULA: 2020XXXX
-> COLUNA 1: BIMESTRE 1
-> COLUNA 2: BIMESTRE 2
"""
import numpy
INICIO = 1000
FIM = 9999
LINHAS = 10 # 10 Alunos
COLUNAS = 4  # Dados da Tabela: 
MEDIA = 6.0
DESVIO = 1.3
pauta = numpy.zeros((LINHAS, COLUNAS))
# Simulação de Dados
pauta[:, 0] = 20200000 + numpy.random.randint(INICIO, FIM + 1, LINHAS, dtype=int)
pauta[:, 1] = numpy.random.normal(MEDIA, DESVIO, LINHAS)
pauta[:, 2] = numpy.random.normal(MEDIA, DESVIO, LINHAS)
pauta
```

#  4: Dataset Academia

```python
"""
-> Dataset Academia: TABELA: LINHAS: 10 x COLUNAS: 5
-> LINHAS: ALUNOS = 10
-> COLUNAS: DADOS DOS ALUNOS
-> COLUNA 0: Matrícula XXX
-> COLUNA 1: Altura
-> COLUNA 2: Massa
-> COLUNA 3: IMC = Massa / (Altura ** 2)
-> COLUNA 4: SEXO: Masculino: 0 ou Feminino: 1
-> COLUNA 5: CATEGORIA:
  -> Categoria 1:
  -> Categoria 2:
  ...
  -> Categoria 6: 

"""
LINHAS = 100  # ALUNOS
COLUNAS = 6
academia = numpy.zeros((LINHAS, COLUNAS)) # MATRIZ(2D): (LINHAS, COLUNAS)
academia[:, 0] = numpy.random.randint(100, 1000, LINHAS)
academia[:, 1] = numpy.random.normal(1.78, 0.15, LINHAS)
academia[:, 2] = numpy.random.normal(78.000, 15.525, LINHAS)
academia[:, 3] = academia[:, 2] / (academia[:, 1] ** 2)     # Vetorização
academia[:, 4] = numpy.random.randint(0, 2, LINHAS)
# Categorização: 1 até 6. (Veja a Tabela)
# Categoria 1 (Indexação Booleana): Abaixo do peso.
indexCategoria1 = academia[:, 3] < 18.5   
academia[indexCategoria1, 5] = 1
# Categoria 2 (Indexação Booleana): Peso Normal.
indexCategoria2 = (academia[:, 3] >= 18.5) & (academia[:, 3] < 25.0)
academia[indexCategoria2, 5] = 2
# Categoria 3 (Indexação Booleana): Sobrepeso.
indexCategoria3 = (academia[:, 3] >= 25.0) & (academia[:, 3] < 30.0)
academia[indexCategoria3, 5] = 3
# Categoria 4 (Indexação Booleana): Obesidade grau I.
indexCategoria4 = (academia[:, 3] >= 30.0) & (academia[:, 3] < 35.0)
academia[indexCategoria4, 5] = 4
# Categoria 5 (Indexação Booleana): Obesidade grau II.
indexCategoria5 = (academia[:, 3] >= 35.0) & (academia[:, 3] < 40.0)
academia[indexCategoria5, 5] = 5
# Categoria 6 (Indexação Booleana): Obesidade grau III.
indexCategoria6 = academia[:, 3] >= 40.0
academia[indexCategoria6, 5] = 6
```

# Vetorização - Exemplo 3 - Dataset Pauta

- TABELA (Matriz): Modelagem do problema
- LINHAS: Eventos (pessoas, mercadorias, ...)
- COLUNAS: Variáveis
- TABELA = LINHAS x COLUNAS = 100 x 4 = 400 Dados
- COLUNA 0: MATRICULA: 2020XXXX
- COLUNA 1: BIMESTRE 1
- COLUNA 2: BIMESTRE 2

```python
import numpy
INICIO = 1000
FIM = 9999
LINHAS = 10 # 10 Alunos
COLUNAS = 4  # Dados da Tabela: 
MEDIA = 6.0
DESVIO = 1.3
pauta = numpy.zeros((LINHAS, COLUNAS))
# Simulação de Dados
pauta[:, 0] = 20200000 + numpy.random.randint(INICIO, FIM + 1, LINHAS, dtype=int)
pauta[:, 1] = numpy.random.normal(MEDIA, DESVIO, LINHAS)
pauta[:, 2] = numpy.random.normal(MEDIA, DESVIO, LINHAS)
pauta
```

```python
# Matrícula dos 5 primeiros alunos da pauta.
pauta[:5,0]
```

```python
# Matrícula dos 5 primeiros alunos da pauta + 1
pauta[:5,0] = pauta[:5,0] + 1                  # Vetorização
pauta[:5,0]
```

```python
# Cálculo da Média Semestral
pauta[:, 3] = (pauta[:, 1] + pauta[:, 2]) / 2  # Vetorização
print('MÉDIA SEMESTRAL:')
for index in range(LINHAS):
  print(f'Aluno {index + 1} - {pauta[index, 0]: .0f}: {pauta[index, 3]: .1f}')
```

# Enumerar: comando enumerate

Comando enumerate = mention (a number of things) one by one.

```python
print('Dados da Pauta: ')
for index, aluno in enumerate (pauta):
  print(f'Dados do aluno {index + 1}')
  print(f'Matricula.     : {aluno[0]: .0f}')
  print(f'Bimestre 1.    : {aluno[1]: .1f}')
  print(f'Bimestre 2     : {aluno[2]: .1f}')
  print(f'Média Semestral: {aluno[3]: .1f}\n')      # Pular linha: \n
```
