```python
import seaborn as snsvariety
import pandas as pd
```

```python
iris = pd.read_csv('C:/Users/tarsi/OneDrive/Documentos/Data Science/Data-Science/Python/Data/iris.csv')
iris.head()
```

## Gráficos de Dispersão

```python
# Relational plot
sns.relplot(x="sepal.length", y="sepal.width", data=iris);
```

```python
# muda cor de acordo com grupo (especie)
sns.relplot(x="sepal.length", y="sepal.width", hue='variety' ,data=iris);
```

```python
# muda estilo do ponto de acordo com grupo (especie), ajuda a enfatizar
sns.relplot(x="sepal.length", y="sepal.width", hue='variety' , style='variety',data=iris);
```

```python
# muda cor de acordo com numero
sns.relplot(x="sepal.length", y="sepal.width", hue='petal.length',data=iris);
```

```python
# muda tamanho do ponto de acordo com tamanho da petala
sns.relplot(x="sepal.length", y="sepal.width", size='petal.length',data=iris);
```
