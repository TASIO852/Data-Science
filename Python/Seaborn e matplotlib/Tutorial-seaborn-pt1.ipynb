```python
import seaborn as sns
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

## Gráficos de Dispersão

```python
# Relational plot
sns.relplot(x="SepalLengthCm", y="SepalWidthCm", data=iris);
```

```python
# muda cor de acordo com grupo (especie)
sns.relplot(x="SepalLengthCm", y="SepalWidthCm", hue='Species' ,data=iris);
```

```python
# muda estilo do ponto de acordo com grupo (especie), ajuda a enfatizar
sns.relplot(x="SepalLengthCm", y="SepalWidthCm", hue='Species' , style='Species',data=iris);
```

```python
# muda cor de acordo com numero
sns.relplot(x="SepalLengthCm", y="SepalWidthCm", hue='PetalLengthCm',data=iris);
```

```python
# muda tamanho do ponto de acordo com tamanho da petala
sns.relplot(x="SepalLengthCm", y="SepalWidthCm", size='PetalLengthCm',data=iris);
```
